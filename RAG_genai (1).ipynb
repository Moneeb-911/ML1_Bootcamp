```python
import torch
import langchain
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
# from transformers import AutoTokenizer, Pipeline
from langchain import HuggingFacePipeline
from langchain.chains import RetrievalQA
from langchain import PromptTemplate
from langchain.document_loaders import PyPDFDirectoryLoader
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
loader=PyPDFDirectoryLoader("/content/data")
data=loader.load()
# print(data)
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=20)
text_chunks=text_splitter.split_documents(data)
```

```python
modelPath='sentence-transformers/all-MiniLM-l6-v2'
model_kwargs={'device':'cpu'}
encode_kwargs={'normalize_embeddings':False}
embeddings=HuggingFaceEmbeddings(
    model_name=modelPath,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
db=FAISS.from_documents(text_chunks,embeddings)
```

```python
db.save_local("faiss")
```

```python
question="what are core skills for AI engineer roadmap to master these skills"
search_docs=db.similarity_search(question)
print(search_docs[0])
```

```python
# llm=CTransformers(
#     model=model_path,
#     model_type='llama',
#     config={'max_new_tokens':128,'temperature':0.1}
# )
llm = CTransformers(
    model=model_path,
    model_type='llama',
    config={'max_new_tokens': 128, 'temperature': 1}
)
```

```python
prompt_template="""Use the following pieces of information to answer the user's question.
If you don't know the answer, just say that you don't know, don't try to make up an answer.
Context:{context}
Question:{question}"""
```

```python
prompt=PromptTemplate(template=prompt_template,input_variables=['context','question'])
chain_type_kwargs={'prompt':prompt}
```

```python
ret=RetrievalQA.from_chain_type(
    llm=llm,
    chain_type='stuff',
    retriever=db.as_retriever(search_kwargs={'k':2}),
    chain_type_kwargs=chain_type_kwargs,
    return_source_documents=True,
    verbose=True
)
```

```python
res=ret.invoke("Explain the Pakistan army")
```

```python
context = res['source_documents']
answer = res['result']

# If context is not relevant or answer is empty, return "I don't know"
if not context or not answer.strip():
    print("I dont know")
else:
    print(answer)
```
